# Criando as tabelas dimensão

```python
# Criação da tabela df_tabela_timeV2 com todos os times que participaram do campeonato brasileiro
tabela_time = set(dados['time_man'])
df_tabela_time = pd.DataFrame(tabela_time)

df_tabela_timeV2 = df_tabela_time.rename(columns={0:'Times'})
```

```python
df_tabela_timeV2
```

```python
lista = np.arange(len(df_tabela_timeV2['Times']))
```

```python
df_tabela_timeV2.insert(1,"indice",lista)
```

```python
#Tabela dimensão times
df_tabela_timeV2.head(50)
```

```python
tabela_arbitro = set(dados['arbitro'])
df_tabela_arbitro = pd.DataFrame(tabela_arbitro)
df_tabela_arbitro = df_tabela_arbitro.dropna(axis=0, how='all')
```

```python
df_tabela_arbitro = df_tabela_arbitro.rename(columns={0:'Arbitro'})
lista_arbitros = np.arange(len(df_tabela_arbitro['Arbitro']))
df_tabela_arbitro.insert(1,"indice",lista_arbitros)
```

```python
df_tabela_arbitro
```

```python
tabela_estadio = set(dados['estadio'])
df_tabela_estadio = pd.DataFrame(tabela_estadio)
df_tabela_estadio = df_tabela_estadio.dropna(axis=0, how='all')
```

```python
df_tabela_estadio
```

```python
df_tabela_estadio = df_tabela_estadio.rename(columns={0:'Estadios'})
lista_estadios = np.arange(len(df_tabela_estadio['Estadios']))
df_tabela_estadio.insert(1,"indice",lista_estadios)
```
